```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("C:\\Users\\HP\\Downloads\\archive\\Mall_Customers.csv")  # Example file name
```

```python
# Map cluster labels to meaningful names
cluster_names = {
    0: "Young High Spenders",
    1: "Older Moderate Earners",
    2: "Middle-Aged Affluent Shoppers",
    3: "Older Low Spenders"
}

df['Cluster Name'] = df['Cluster'].map(cluster_names)

# Display the updated DataFrame
print(df.head())
```

```python
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
```

```python
features = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Gender']]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title('Elbow Method to Determine Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_features)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=pca_components[:, 0],
    y=pca_components[:, 1],
    hue=df['Cluster'],
    palette='Set2',
    s=100
)
plt.title('Customer Segmentation using K-means Clustering')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster')
plt.show()
```

```python
sil_score = silhouette_score(scaled_features, df['Cluster'])
print(f'Silhouette Score: {sil_score}')
```
